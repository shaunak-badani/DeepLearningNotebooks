```python
chars = sorted(list(set(text)))
vocab_size = len(chars)
print(''.join(chars))
```

## Create a mapping from characters to integers

- Create a mapping that converts characters to integers
- And a reverse mapping for integers to characters
- The index of each character in the sorted array is the integer value for that character

```python
stoi = { ch:i for i, ch in enumerate(chars) }
itos = { i:ch for i, ch in enumerate(chars) }

def encode(s):
    encoded_string = []
    for c in s:
        encoded_string.append(stoi[c])
    return encoded_string

def decode(s):
    decoded_string = []
    for c in s:
        decoded_string.append(chars[c])
    return ''.join(decoded_string)

print("Encoding of Hii there! :")
print(encode("Hii there!"))
print("Decoding of above array : ")
print(decode(encode("Hii there!")))
```

```python
import torch

data = torch.tensor(encode(text), dtype = torch.long)
print(data.shape, data.dtype)
```

```python
# Splitting data into train and validation sets
n = int(0.9 * len(data))
train_data = data[:n]
val_data = data[n:]
```

```python
block_size = 8
train_data[:block_size + 1]
```

```python
x = train_data[:block_size]
# x is an array of size (block_size)
y = train_data[1:block_size+1]
# y is the same array offset by 1, size (block_size)
for t in range(block_size):
    context = x[:t+1]
    # context is a (t + 1) length vector
    target = y[t]
    print(f"When the input is {context}, the target is {target}")
```

```python
len(data)
```

```python
torch.manual_seed(1337)
batch_size = 4
block_size = 8

def get_batch(split):
    """
    Generate a small batch of data of inputs x and targets y
    """
    data = train_data if split == "train" else val_data
    ix = torch.randint(len(data) - block_size, (batch_size, ))
    # ix are indices
    x = torch.stack([data[i:i+block_size] for i in ix])
    y = torch.stack([data[i + 1:i + block_size + 1] for i in ix])
    return x,y

xb,yb = get_batch("train")
print("inputs:")
print(xb.shape)
print(xb)
print("targets")
print(yb.shape)
print(yb)
```

```python
import torch
import torch.nn as nn
from torch.nn import functional as F

channel_size = vocab_size
torch.manual_seed(1337)

class BigramLanguageModel(nn.Module):

    def __init__(self, vocab_size):
        super().__init__()
        self.token_embedding_table = nn.Embedding(vocab_size, channel_size)
        
    def forward(self, idx, targets=None):
        logits = self.token_embedding_table(idx) # dimensions (Batch_size, time, Channels = channel_size)

        if targets is None:
            loss = None
        else:
            B, T, C = logits.shape
            logits = logits.view(B * T, C)
            targets = targets.view(B * T)
            loss = F.cross_entropy(logits, targets)
        # For each token, we get a (channel_size) vector which is the probability of the next token being i. i belongs to [0, vocab_size)
        return logits, loss

    def generate(self, idx, max_new_tokens):
        for _ in range(max_new_tokens):
            logits, loss = self(idx)
            logits = logits[:, -1, :]

            probs = F.softmax(logits, dim = 1)
            idx_next = torch.multinomial(probs, num_samples = 1)
            idx = torch.cat((idx, idx_next), dim = 1)
        return idx

m = BigramLanguageModel(vocab_size)
out, loss = m(xb, yb)
```

### Mathematical trick used in self attention

```python
torch.manual_seed(1337)

B, T, C = 4, 8, 2
x = torch.randn(B, T, C)
x.shape
```

```python
# we want x[b, t] = mean_{i <= t} x[b, i]
xbow = torch.zeros((B, T, C))
for b in range(B):
    for t in range(T):
        xbow[b, t] = torch.mean(x[b, :t+1], 0, keepdims = True)
        
```

```python
xbow.shape
```

```python
wei = torch.tril(torch.ones((T, T)))
wei = wei / wei.sum(1, keepdims = True)
wei
```

```python
wei.shape
```

```python
xbow2 = wei @ x
```

```python
torch.allclose(xbow, xbow2, 1e-4)
```

```python
tril = torch.tril(torch.ones((T, T)))
wei = torch.zeros((T, T))
wei = torch.masked_fill(wei, tril == 0, float('-inf'))
wei = F.softmax(wei, dim = 1)
xbow3 = wei @ x
```

```python
torch.allclose(xbow3, xbow, 1e-3)
```

```python
import torch
import torch.nn as nn
from torch.nn import functional as F

n_embd = 32
channel_size = vocab_size
torch.manual_seed(1337)

class BigramLanguageModel(nn.Module):

    def __init__(self):
        super().__init__()
        self.token_embedding_table = nn.Embedding(vocab_size, n_embd)
        self.position_embedding_table = nn.Embedding(vocab_size, n_embd)
        self.lm_head = nn.Linear(n_embd, vocab_size)

    def forward(self, idx, targets=None):
        B, T = idx.shape
        tok_embd = self.token_embedding_table(idx) # dimensions (B, T, C)
        pos_embd = self.position_embedding_table(torch.arange(T)) # (T , C)
        x = tok_embd + pos_embd
        logits = self.lm_head(x)

        if targets is None:
            loss = None
        else:
            B, T, C = logits.shape
            logits = logits.view(B * T, C)
            targets = targets.view(B * T)
            loss = F.cross_entropy(logits, targets)
        # For each token, we get a (channel_size) vector which is the probability of the next token being i. i belongs to [0, vocab_size)
        return logits, loss

    def generate(self, idx, max_new_tokens):
        for _ in range(max_new_tokens):
            idx_cond = idx[:, -block_size:]
            logits, loss = self(idx_cond)
            logits = logits[:, -1, :]

            probs = F.softmax(logits, dim = 1)
            idx_next = torch.multinomial(probs, num_samples = 1)
            idx = torch.cat((idx, idx_next), dim = 1)
        return idx

m = BigramLanguageModel()
out, loss = m(xb, yb)
optimizer = torch.optim.AdamW(m.parameters(), lr = 1e-3)
```

### Version 4: Self attention!

```python
torch.manual_seed(1337)
B,T,C = 4, 8, 32
x = torch.randn(B, T, C)
print(x.shape)
head_size = 16
key = nn.Linear(C, head_size, bias = False)
query = nn.Linear(C, head_size, bias = False)
value = nn.Linear(C, head_size, bias = False)
k = key(x) # (B, T, 16)
q = query(x) # (B, T, 16)
wei = q @ k.transpose(-2, -1) # (B, T, 16 ) @ (B, 16 T) => (B, T, T)
v = value(x)
out = wei @ v

wei = wei.masked_fill(torch.tril(torch.ones(T, T)) == 0, float('-inf'))
wei = F.softmax(wei, dim = -1) # -1 means the last dimension

out.shape
```

### Coding up the head

```python
class Head(nn.Module):
    """One head of self attention"""

    def __init__(self, head_size):
        super().__init__()
        self.key = nn.Linear(n_embd, head_size, bias = False)
        self.query = nn.Linear(n_embd, head_size, bias = False)
        self.value = nn.Linear(n_embd, head_size, bias = False)
        self.register_buffer('tril', torch.tril(torch.ones(block_size, block_size)))

    def forward(self, x):
        B, T, C = x.shape
        k = self.key(x) # (B, T, 16)
        q = self.query(x) # (B, T, 16)
        v = self.value(x) # (B, T, 16)

        wei = q @ k.transpose(-2, -1) * head_size ** (-0.5)
        wei = wei.masked_fill(self.tril == 0, float('-inf'))
        wei = F.softmax(wei, dim = -1)
        out = wei @ v
        return out

class BigramLanguageModel(nn.Module):

    def __init__(self):
        super().__init__()
        self.token_embedding_table = nn.Embedding(vocab_size, n_embd)
        self.position_embedding_table = nn.Embedding(vocab_size, n_embd)
        self.sa_head = Head(n_embd)
        self.lm_head = nn.Linear(n_embd, vocab_size)

    def forward(self, idx, targets=None):
        B, T = idx.shape
        tok_embd = self.token_embedding_table(idx) # dimensions (B, T, C)
        pos_embd = self.position_embedding_table(torch.arange(T)) # (T , C)
        x = tok_embd + pos_embd
        x = self.sa_head(x)
        logits = self.lm_head(x)

        if targets is None:
            loss = None
        else:
            B, T, C = logits.shape
            logits = logits.view(B * T, C)
            targets = targets.view(B * T)
            loss = F.cross_entropy(logits, targets)
        # For each token, we get a (channel_size) vector which is the probability of the next token being i. i belongs to [0, vocab_size)
        return logits, loss

    def generate(self, idx, max_new_tokens):
        for _ in range(max_new_tokens):
            idx_cond = idx[:, -block_size:]
            logits, loss = self(idx_cond)
            logits = logits[:, -1, :]

            probs = F.softmax(logits, dim = 1)
            idx_next = torch.multinomial(probs, num_samples = 1)
            idx = torch.cat((idx, idx_next), dim = 1)
        return idx

m = BigramLanguageModel()
out, loss = m(xb, yb)
optimizer = torch.optim.AdamW(m.parameters(), lr = 1e-3)

```
